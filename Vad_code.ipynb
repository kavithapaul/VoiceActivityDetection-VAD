```python
from pydub import AudioSegment, silence

def split_audio_on_silence(filename, min_silence_len, silence_thresh):
    """
    Splits the audio file into chunks based on silence.

    :param filename: Path to the input WAV file.
    :param min_silence_len: Minimum length of silence in milliseconds to consider for splitting.
    :param silence_thresh: Silence threshold (in dB).
    :return: List of audio chunks.
    """
    # Load the audio file
    audio = AudioSegment.from_file(filename, format="wav")

    # Split on silence
    chunks = silence.split_on_silence(audio,
                                      min_silence_len=min_silence_len,
                                      silence_thresh=silence_thresh)

    # Export the chunks as separate audio files
    for i, chunk in enumerate(chunks):
        chunk_filename = f"chunk_{i}.wav"
        chunk.export(chunk_filename, format="wav")
        print(f"Exported {chunk_filename}")

# Adjust parameters as needed
split_audio_on_silence('/content/sample_data/Speech.wav', min_silence_len=900, silence_thresh=-45)
```
